```python
import requests
import json
```

```python
query=input('What Type of news are you interested in ?')
url=f"https://newsapi.org/v2/everything?q={query},&from=2024-08-19&sortBy=publishedAt&apiKey=3f820bd5857d495ab4f5c1cf8afb92de"
result=requests.get(url)
news=json.loads(result.text)
for article in news['articles']:
  print(article['title'])
  print(article['description'])
  print('--------------------------------------')
```
